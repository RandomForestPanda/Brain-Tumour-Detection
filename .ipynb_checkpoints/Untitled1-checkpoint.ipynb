# Import packages

```python
import numpy as np
import torch  
from torch.utils.data import Dataset, DataLoader  
import glob                                     # to navigate thru different packages 
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix,accuracy_score     # confsuion matrix is used to represent different combinations possible out of the classes of objects present
import cv2
```

## Reading the images

```python
tumor = []
healthy=[]


path_tumor= './data/brain_tumor_dataset/yes/*.jpg'  #in current directory go to data ->brain-tumor_dataset->yes and all the .jpg file we find   
for f in glob.iglob(path_tumor):   #iglob takes the path of concern 
    img = cv2.imread(f)
    img=cv2.resize(img,(128,128))     # the input images should be consistent in width and height
    b ,g ,r=cv2.split(img)            #split the channels 
    img=cv2.merge([r,g,b])            #merge the channels such that RGB format is followed 
    tumor.append(img)                 # all three channels are of same dmensions that is 128 x 128




path_healthy='./data/brain_tumor_dataset/no/*.jpg'
for f in glob.iglob(path_healthy):
    img=cv2.imread(f)
    img=cv2.resize(img,(128,128))
    b,g,r=cv2.split(img)
    img=cv2.merge([r,g,b])
    healthy.append(img)
```

# VISUALIZING BRAIN TUMOR IMAGES

```python
# plt.imshow(healthy[0])
```

```python
def plot_random(healthy,tumor,num=5):
    # for i in range(5):
    healthy_images=healthy[np.random.choice(healthy.shape[0],num,replace=False)]
    tumor_images=tumor[np.random.choice(tumor.shape[0],num,replace=False)]
    plt.figure(figsize=(16,9))
    for i in range(num):
        plt.subplot(1,num,i+1)
        plt.imshow(healthy_images[i])
    
    for i in range(num):
        plt.subplot(1,num,i+1)
        plt.imshow(tumor_images[i])
        
    
```

```python
plot_random(healthy,tumor)
```
