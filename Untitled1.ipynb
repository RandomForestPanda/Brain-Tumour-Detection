# CREATING MRI CUSTOM DATASET CLASS

```python
class MRI(Dataset):
    def __init__(self):
        tumor = []
        healthy=[]


        path_tumor= './data/brain_tumor_dataset/yes/*.jpg'  #in current directory go to data ->brain-tumor_dataset->yes and all the .jpg file we find   
        for f in glob.iglob(path_tumor):   #iglob takes the path of concern 
                img = cv2.imread(f)
                img=cv2.resize(img,(128,128))     # the input images should be consistent in width and height
                b,g,r=cv2.split(img)            #split the channels 
                img=cv2.merge([r,g,b])            #merge the channels such that RGB format is followed 
                img=img.reshape((img.shape[2],img.shape[0],img.shape[1]))
                tumor.append(img)                 # all three channels are of same dmensions that is 128 x 128




        path_healthy='./data/brain_tumor_dataset/no/*.jpg'
        for f in glob.iglob(path_healthy):
                img=cv2.imread(f)
                img=cv2.resize(img,(128,128))
                b,g,r=cv2.split(img)
                img=cv2.merge([r,g,b])
                img=img.reshape((img.shape[2],img.shape[0],img.shape[1]))
                healthy.append(img)
        #images
        tumor=np.array(tumor,dtype=np.float32)
        healthy=np.array(healthy,dtype=np.float32)

        #labels
        tumor_label=np.ones(tumor.shape[0],dtype=np.float32)
        healthy_label=np.zeros(healthy.shape[0],dtype=np.float32) 

        self.images=np.concatenate((tumor,healthy),axis=0)  # images and labels are put into two variables
        self.labels=np.concatenate((tumor_label,healthy_label))
        
    def __len__(self):
        return self.images.shape[0]
    def __getitem__(self,index):
        """
        returns a dictionary containing the image and the corresponding label 
        """
        sample={'image':self.images[index],'label':self.labels[index]}
        return sample
    #adding more functionalities  for pre processing 
    def normalize(self):
        self.images=self.images/255.0
    
    
                                                                    
        
        
```

## THE CNN MODEL

```python
import torch.nn as nn
import torch.nn.functional as F

class CNN(nn.Module):

    def __init__(self):
        super(CNN,self).__init__() # super () allows to acess the overridden methods under the class which has overridden it, here CCN can ascess the methods of nn.Module even though CNN overrdies the nn.Module class 
        #IN PYTORCH, can create layers just be making them as an instance in the constructor
        self.cnn_model=nn.Sequential(
        nn.Conv2d(in_channels=3,out_channels=6,kernel_size=(5,5)), # learns detection of basic lines,shapes
        nn.Tanh(), #no args, its a mathematical method that will be implemented in each of the 6 out_channels
        nn.AvgPool2d(stride=5,kernel_size=2),
        nn.Conv2d(in_channels=6,out_channels=16,kernel_size=(5,5)), #learns more detailed features
        nn.Tanh(),
        nn.AvgPool2d(stride=5,kernel_size=2))
            # nn.Flatten(

        self.fc_model=nn.Sequential(
        nn.Linear(in_features=256 ,out_features=120),
        nn.Tanh(),
        nn.Linear(in_features=120,out_features=84),
        nn.Tanh(),
        nn.Linear(in_features=84,out_features=1),
            # nn.Tanh(),
            # nn.Linear(in_features=64,out_features=1)


        )

    def forward(self,x):
        x=self.cnn_model(x)
        x=x.view(x.size(0),-1) #flattens the 2d arrays
        x=self.fc_model(x)
        x=F.sigmoid(x)
        return x
```

```python
model=CNN().to(mps_device)

dataloader=DataLoader(mri_dataset,batch_size=32,shuffle=False)
```

```python
model.eval()
outputs=[]
y_true=[]
# by default everything is on cpu hence move the dataset onto gpu using .to(mps_device)
with torch.no_grad():
    for D in dataloader:
        image=D['image'].to(mps_device)
        label=D['label'].to(mps_device) #just put the images of the dataset onto gpu no need to put the ground truth labels onto gpu
        y_hat=model(image)  #this is a gpu tensor
        
        #convert the gpu tensor into a cpu tensor by detaching and then into a numpy array 
        outputs.append(y_hat.cpu().detach().numpy()) 
        
        y_true.append(label.cpu().detach().numpy())
        # print(f"predictions{outputs}   ground_truth {y_true}")
    
```

```python
outputs=np.concatenate(outputs,axis=0).squeeze()
print(outputs)
y_true=np.concatenate(y_true,axis=0).squeeze()
print(len(y_true)+50)
```

```python
def threshold(scores,threshold=0.5,minimum=0,maximum=1.0):
    x=np.array(list(scores))
    
    x[x>=threshold]=maximum
    x[x<threshold]=minimum
    return x
```

# TRAINING MODEL

```python
eta=0.0001 #
epochs=300
optimizer=torch.optim.Adam(model.parameters(),lr=eta)
dataloader=DataLoader(mri_dataset,batch_size=32,shuffle=True)
model.train()
```

```python

for epoch in range(1,epochs):
    losses=[]
    # print(type(losses))
    for D in dataloader :
        optimizer.zero_grad ()  #before feeding the next data points make sure to zero all the gradients
        data=D['image'].to(mps_device)
        label=D['label'].to(mps_device)
        y_hat=model(data)
        # y_hat=y_hat.squeeze()  #uncomment to make dimension of y_hat == dimension of label
        
        # print(y_hat.shape,label.shape)
        # sys.exit()
        error=nn.BCELoss()
        loss=torch.sum(error(y_hat.squeeze(),label))
        loss.backward() #backtrack
        optimizer.step() 
        losses.append(loss.cpu().detach().numpy())
    # losses.to(mps_device)
    # print(type(losses))
    if (epoch-1)%10==0:
        print("train epoch: {}\tLoss: {:.6f}".format(epoch+1,np.mean(losses)))
        
```

```python
model.eval()
outputs=[]
y_true=[]
# by default everything is on cpu hence move the dataset onto gpu using .to(mps_device)
with torch.no_grad():
    for D in dataloader:
        image=D['image'].to(mps_device)
        label=D['label'].to(mps_device) #just put the images of the dataset onto gpu no need to put the ground truth labels onto gpu
        y_hat=model(image)  #this is a gpu tensor
        
        #convert the gpu tensor into a cpu tensor by detaching and then into a numpy array 
        outputs.append(y_hat.cpu().detach().numpy()) 
        
        y_true.append(label.cpu().detach().numpy())
        # print(f"predictions{outputs}   ground_truth {y_true}")
    
outputs=np.concatenate(outputs,axis=0).squeeze()
y_true=np.concatenate(y_true,axis=0).squeeze()
```
